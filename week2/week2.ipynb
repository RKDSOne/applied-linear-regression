```python
from sklearn.linear_model import LinearRegression
import numpy as np
import scipy as sp
import statsmodels.api as sm
```

```python
X = df[["SMK"]].values
X = sm.add_constant(X)  # add the intercept term
y = df["SBP"].values
ols = sm.OLS(y, X).fit()
ols.summary()
```

```python
df.describe()
```

```python
ggplot(df, aes('SMK','SBP')) + \
     geom_point(colour="black", size = 100) +\
    ggtitle("SBP (Y) vs. SMK (X)") + \
    geom_line() +\
    stat_smooth(colour='blue', span=0.2)
```

```python
reg = LinearRegression()
X = df[["SMK"]].values
y = df["SBP"].values
coefficients = reg.fit(X, y)
(reg.intercept_, reg.coef_)
```

```python
ggplot(df, aes('SMK','SBP')) + \
     geom_point(colour="black", size = 100) +  geom_abline(intercept=reg.intercept_,slope=reg.coef_[0],
                colour="red",size=2) + ggtitle("Regression SBP (Y) vs. SMK (X)")
```
